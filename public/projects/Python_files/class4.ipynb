```python
import numpy as np
a1=np.array([12,32,48,47,54,98,-24,12])
print(np.min(a1))
print(np.max(a1))
print(np.sum(a1))
print(np.product(a1))
```

```python
import numpy as pd
import matplotlib.pyplot as plt
x=np.linspace(0,10,100)
y=x
plt.plot(x,y)
plt.show()
x=np.linspace(-10,10,100)
y=x**2
plt.plot(x,y)
plt.show()
```

```python
x=np.linspace(0,10,100)
y=np.sin(x)
plt.plot(x,y)
plt.show()
x=np.linspace(-10,10,100)
y=1/(1+np.exp(-x))
plt.plot(x,y)
plt.show()
```

```python
v=np.array([1,2,3])
norm=np.linalg.norm(v)
print(norm)
```

```python
v1=np.array([1,2,3])
v2=np.array([4,5,6])
v_sum=np.add(v1,v2)
print(v_sum)
```

```python
v1=np.array([1,2,3])
v2=np.array([4,5,6])
dot_product=np.dot(v1,v2)
print(dot_product)
```

```python
matrix1=np.array([[1,3],
                 [5,7]])
matrix2=np.array([[2,6],
                  [4,8]])
result_matrix=np.add(matrix1,matrix2)
print(result_matrix)
```

```python
v1=np.array([1,2,3])
v2=np.array([4,5,6])
dot_product=np.dot(v1,v2)
print(dot_product)
```

```python
matrix1=np.array([[1,2],[3,4]])
matrix2=np.array([[5,6],[7,8]])
result_matrix1=np.matmul(matrix1,matrix2)
print(result_matrix1)
```

```python
matrix1=np.array([[1,3],
                  [5,7]])
print(np.transpose(matrix1))
```

```python
matrix1=np.array([[1,3,5],[7,9,2],[4,6,8]])
result2=np.linalg.inv(matrix1)
print(result2)
```

```python
matrix3=np.array([[1,2,3],
                  [4,5,1],
                  [2,3,4]])
result3=np.linalg.det(matrix3)
print(result3)
```
